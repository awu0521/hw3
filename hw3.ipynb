Answer each of the following questions in 2 to 3 sentences. 

1. What is the problem with calling `fit_transform` on your test data with `StandardScaler`?
2. Why is it important to follow the Golden Rule? If you violate it, will that give you a worse classifier?
3. What are two advantages of using sklearn Pipelines? 
4. When is it appropriate to use sklearn `ColumnTransformer`? 

<div class="alert alert-warning">
    
Solution_4.5

_Points:_ 8

1. Calling fit_transform on the test data would cause the scaler to learn the mean and standard deviation from the test set and use those to scale it. This is a problem because the scaler should only be fit on the training data, and the test set should be transformed using the statistics learned from training, not its own.

2. The Golden Rule states that no information from the test set should leak into the training process. Violating it would actually give you a falsely optimistic classifier, not a worse one, because the model has indirectly seen information about the test data during training. This means your evaluation metrics would overestimate how well the model generalizes to truly unseen data.

3. One advantage of using pipelines is that they prevent data leakage by ensuring that each preprocessing step is fit only on training data and applied consistently to test data. Another advantage is that they make the code cleaner (they make it into one object).

4. When we have different types of features that each require a different preprocessing transformation. For example, if some features need scaling while others need one-hot encoding or need to be dropped, ColumnTransformer lets us apply each transformation to the appropriate subset of columns at the same time.

## Exercise 5: Building models 

Now that we have preprocessed features, we are ready to build models. Below, I'm providing the function we used in class which returns mean cross-validation score along with standard deviation for a given model. Use it to keep track of your results. 

```python
results_dict = {}  # dictionary to store all the results
```

```python
def mean_std_cross_val_scores(model, X_train, y_train, **kwargs):
    """
    Returns mean and std of cross validation

    Parameters
    ----------
    model :
        scikit-learn model
    X_train : numpy array or pandas DataFrame
        X in the training data
    y_train :
        y in the training data

    Returns
    ----------
        pandas Series with mean scores from cross_validation
    """

    scores = cross_validate(model, X_train, y_train, **kwargs)

    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = []

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores.iloc[i], std_scores.iloc[i])))

    return pd.Series(data=out_col, index=mean_scores.index)
```

Below, I'm showing an example where I call `mean_std_cross_val_scores` with `DummyClassifier`. The function calls `cross_validate` with the passed arguments and returns a series with mean cross-validation results and std of cross-validation. When you train new models, you can just add the results of these models in `results_dict`, which can be easily converted to a dataframe so that you can have a table with all your results. 

```python
# Baseline model

from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(random_state = 123)
pipe = make_pipeline(preprocessor, dummy)
results_dict["dummy"] = mean_std_cross_val_scores(
    pipe, X_train, y_train, cv=5, return_train_score=True
)
results_df = pd.DataFrame(results_dict).T
results_df
```

### 7.1 kNNs toy example
rubric={points}

Suppose you want to get predictions using the $k$-nearest neighbour algorithm on the toy dataset below. 

$$ X = \begin{bmatrix}5 & 2\\4 & -2\\  2 & 2\\ 10 & 10\\ 9 & -1\\ 9& 9\end{bmatrix}, \quad y = \begin{bmatrix}0\\0\\1\\1\\1\\2\end{bmatrix}.$$

Suppose you are given a new data point $x=\begin{bmatrix} 0 & 0\end{bmatrix}$.  

1. What would the model predict when $k=1$?
2. What would the model predict when $k=3$?
3. What would the model predict when $k=3$ if we were doing regression rather than classification? 

<div class="alert alert-warning">
    
Solution_7.1
    
</div>

_Points:_ 4

When k = 1, the nearest neighbor is [2,2] with label 1, so the prediction is 1.

When k = 3, the 3 nearest neighbors are [2, 2], [4, -2], and [5, 2] with labels 1, 0, and 0. The majority class is 0, so the prediction is 0.

When k = 3 with regression, instead of taking the majority vote we take the mean of the 3 nearest neighbors' labels: (1 + 0 + 0) / 3 = 0.333.
